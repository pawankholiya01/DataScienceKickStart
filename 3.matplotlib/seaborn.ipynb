```python
import seaborn as sns
import numpy as np
```

```python
print(tips.head())
```

```python
# Named Arguments
#sns.barplot(x =,y = , data = )

# estimator is live average etc.

# hue : catagorical value that futher distributes the plot for eact cat
sns.barplot(x ='sex' ,y ='total_bill' , data = tips,estimator = np.var,hue = 'smoker')


```

**Count Plot :** Just a bar graph with counting as an agregator Function

```python
# it counts hence no need of y
sns.countplot(x = 'smoker',data = tips)
```

```python
sns.boxplot(y='total_bill',x='day',data = tips,hue='smoker')
```

```python
# very similar to box plot
#sns.violinplot(y='total_bill',x='day',data = tips,hue='smoker')
sns.violinplot(y='total_bill',x='day',data = tips,hue='smoker',split=True)
```

### pivot_table
its a pandas function it makes dataframe with variables on both axis **SpreadSheet Style** out of a given dataframe  
**index :** Variable for X axis  
**columns :** Variable for Y axis   
**values :** Variable whose values is to be on cell

```python
flight_pivot = flights.pivot_table(index= 'month',columns = 'year',values='passengers')
print(flight_pivot)
```

```python
sns.heatmap(flight_pivot)
```
