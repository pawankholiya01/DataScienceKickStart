```python
import seaborn as sns
import numpy as np
```

```python
tips = sns.load_dataset('tips')
```

```python
tips
print(type(tips))
```

#### Seaborn Works well with pandas dataframe and also it automatically loads in pandas DataFrame

```python
print(tips.head())
```

```python
# Named Arguments
#sns.barplot(x =,y = , data = )

# estimator is live average etc.

# hue : catagorical value that futher distributes the plot for eact cat
sns.barplot(x ='sex' ,y ='total_bill' , data = tips,estimator = np.var,hue = 'smoker')


```

**Count Plot :** Just a bar graph with counting as an agregator Function

```python
# it counts hence no need of y
sns.countplot(x = 'size',data = tips)
```

```python
sns.boxplot(y='total_bill',x='day',data = tips,hue='smoker')
```

```python
# very similar to box plot
#sns.violinplot(y='total_bill',x='day',data = tips,hue='smoker')
sns.violinplot(y='total_bill',x='day',data = tips,hue='smoker',split=True)
```
