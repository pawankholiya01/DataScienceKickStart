# MatplotLib
## Data Visualization with python

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
print(plt.style.available)
```

```python
plt.style.use("bmh")
```

```python
x = np.arange(10)
y = x**2
z = x**3
#plt.plot(x,z)
plt.plot(x,z,color='green')
plt.xlabel("number")
plt.ylabel("cube")
plt.show()
plt.xlabel("number")
plt.ylabel("square")

plt.plot(x,y,color='red')
plt.show()

plt.title("Blah")
plt.plot(x,z,color='green',label = 'Cube')
plt.plot(x,y,color='red',label = "square")

#so as the labels show up
plt.legend()
plt.show()


```
