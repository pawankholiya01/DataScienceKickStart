```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
X = pd.read_csv('./dataset/Linear_X_Train.csv')
Y = pd.read_csv('./dataset/Linear_Y_Train.csv')
XT  = pd.read_csv('./dataset/Linear_X_Test.csv')
```

```python
plt.scatter(X,Y)
plt.show()
```

```python
X_train = X.values.reshape((-1))
Y_train = Y.values.reshape((-1))
X_test = XT.values.reshape((-1))

N = Y_train.shape[0]
print(Y_train.shape)
```

```python
def predY(m,x,c):
    return m*x+c
```

```python
def error(Y_train,Y_pred):
    err = sum((Y_train-Y_pred)**2)/N
    return err
```

```python
lr = 0.1
m = 0
c = 0
errorlist = []
mlist=[]
clist = []
for i in range(1000):
    Y_pred = predY(m,X_train,c)
    m = m - lr*((sum((Y_pred - Y_train)*X_train))/N)
    c = c - lr*((sum(Y_pred - Y_train))/N)
    mlist.append(m)
    clist.append(c)
    errorlist.append(error(Y_train,Y_pred))
    print(c)

```

```python
print(m,c)
Y_test = predY(m,X_test,c)
df = pd.DataFrame(Y_test,columns = ['y'])
```

### Accuracy on Training Data

```python
Y_tt = predY(m,X_train,c)
```

```python
def r2_score(Y_train,Y_tt):

    r2 = 1-(np.sum((Y_train-Y_tt)**2)/(np.sum((Y_train.mean()-Y_tt)**2)))
    return r2
```

```python
print(r2_score(Y_train,Y_tt))
```
