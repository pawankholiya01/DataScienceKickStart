```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dfx = pd.read_csv("./dataset/xdata.csv",index_col=0)
dfy = pd.read_csv("./dataset/ydata.csv",index_col=0)
print(dfx)
print(dfy)
```

```python
X = dfx.values
Y = dfy.values

Y = Y.reshape((-1)) # 2-D to 1-D
#print(X)
#print(Y)
print(X.shape)
print(Y.shape)
```

```python
plt.scatter(X[:,0],X[:,1],c=Y)
plt.show()
```

```python

query = np.array([3,3])
plt.scatter(X[:,0],X[:,1],c=Y)
plt.scatter(query[0],query[1],color= 'red')
plt.show()
```

```python
def dist(x1,x2):
    return np.sqrt(sum((x1-x2)**2))

def knn(X,Y,queryPoint,k=5):
    vals = []
    n = X.shape[0]

    for i in range(n):
        d = dist(queryPoint,X[i])
        vals.append((d,Y[i]))

    vals.sort()
    #print(vals)
    labels=[]
    for i in range(k):
        d,label = vals[i]
        labels.append(label)
    
    
    labels = np.array(labels)
    unique, counts = np.unique(labels, return_counts=True)
    
    #print(unique)
    #print(counts)

    ind  = np.argmax(counts)  # Returns Index of Maximum value in the Array

    return unique[ind]



queryX = np.array([2,2.5])
queryY = knn(X,Y,queryX,5)
print(f"Class of the Given point is {queryY}")



```

```python
# Testing Our function on all of the Test Data

n = X.shape[0]
resY = []
for i in range(n):

    temp =  knn(X,Y,X[i],5)

    resY.append(temp)

z = Y == resY
print((z))

print(np.count_nonzero(Y==resY))
```
