## Scatter Plots

```python
plt.scatter(x,z,color='green',label = 'Cube',marker='o')
plt.scatter(x,y,color='red',label = "square",marker='*')
plt.legend()
plt.title("Scatter Plots")
plt.show()
```

```python
#Adjust Size of Any plot
plt.figure(figsize=(10,5))

plt.scatter(x,z,color='green',label = 'Cube',marker='o')
plt.scatter(x,y,color='red',label = "square",marker='*')
plt.legend()
plt.title("Scatter Plots")
plt.show()
```

## Bar Graphs

```python

x = np.array([0,1,2])
y = np.array([5,2,6])
z = np.array([3,7,4])
plt.bar(x,height=z)
plt.bar(x,height=y)

plt.plot()
#plt.bar?
```

```python
# to prevent overlapping we displace x a little bit for one of them
plt.figure(figsize=(10,5))
plt.bar(x*2 -0.25,height=z,width=0.5,label = "current",tick_label=['x','y','z'])
plt.bar(x*2 + 0.25,height=y,width=0.5,label = "next")
plt.legend()
plt.xlabel("Winter is comming")
plt.title("Bar Graph")
plt.ylim(0,15)
plt.plot()
```

## Pie Charts

```python
skills = "CP","FrontEnd","Backend"
weight = [25,10,20]
plt.pie(weight,labels=skills)

plt.show()
```

```python
#explode : used to highlight specifies radiums

plt.figure(figsize=(7,7))

skills = "CP","FrontEnd","Backend"
weight = [25,10,20]
plt.pie(weight,labels=skills,explode=(0,0.1,0),autopct="%1.1f%%",shadow=True)

plt.show()
```

## Histogram
StandardNormal Distribution


```python
# Generating Stardard Normal Distribution with mean and simga SD
X = np.random.randn(100)
mean = 60
sigma = 5
x = X*sigma + mean
x2 = np.round(X*5+55)

#converting into int
x = np.round(x)
print(x)
```

```python
plt.scatter(x,np.arange(100))
plt.show()
```

Data is more accumulated towards 60 which is the mean and more specifically 60+- 5 which is the Standard Deviation

```python
plt.style.use("seaborn")
plt.hist(x,alpha=.8,LABEL="Alpha")
plt.hist(x2,alpha=.9,label = "Beta")
plt.xlabel("Marks")
plt.ylabel("Freq")
plt.legend()
plt.show()
```

## Data Visualization of Movie DataSet

```python
import pandas as pd
```

```python
df= pd.read_csv("./movies_metadata.csv")
```

```python
df.head(n = 10)
```

```python
df.columns
```

```python
# df.columnname
# df.get('columnname') typesafe does not throw error
# df['columnname]

titles = df.original_title
print((titles))
```

```python
titles = list(titles)
print(type(titles))
```

```python
titles[:5]
```

```python
freq_titles = {}
for title in titles:
    l = len(title)

    if freq_titles.get(l) is None:
        freq_titles[l] = 1
    else:
        freq_titles[l] += 1

print(freq_titles)
        
```

```python
x = np.array(list(freq_titles.keys()))
y = np.array(list(freq_titles.values()))
print(x)
print(y)

```

```python
plt.scatter(x,y,marker="o")
plt.xlabel("length of Titles")
plt.ylabel("Frequency")
plt.title("Movie Data Visualtisation")
plt.show()
plt.hist(y)
plt.show()
```
