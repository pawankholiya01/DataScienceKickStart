## Gradient Descent(General Approach)

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.arange(10)
print(X)
Y = (X-5)**2
print(Y)
```

### Given a function f(X) find Value of X(local-minima) that minimizes the F(X)

```python
plt.style.use('seaborn')
plt.scatter(y= Y,x=X)
plt.show() 
```

```python

plt.plot(X,Y)
plt.show() 
```

```python
# initialise x axis with any random value

x = 0
lr = 0.1

# Let say we will take 50 steps downhill
arr = []
error = []
for i in range(50):
    grad = 2*(x-5)
    x = x - lr*(grad)
    print(x)
    error.append((x-5)**2)
    arr.append(x)


print(x)
```

```python
plt.plot(range(50),error)
plt.title('Error vs Iterations')
plt.show()
plt.title('X vs Iterations')
plt.plot(range(50),arr)
plt.show()
plt.title('Error vs X')
plt.plot(X,Y)
plt.plot(arr,error,color='red')
plt.show()
```
